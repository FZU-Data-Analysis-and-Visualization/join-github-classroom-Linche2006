### 一、基础创建与查看

1. 导入 pandas，用字典创建 DataFrame：

### 三、缺失值

1. 创建含缺失值的 DataFrame：

   data = {"x":[1, None, 3], "y":[None, 5, 6]}

   检测缺失值并统计每列缺失数量

```python
# ====================== 三、缺失值 ======================
print("\n" + "="*50)
print("三、缺失值")
print("="*50)

# 1. 创建含缺失值的 DataFrame，检测缺失值并统计每列缺失数量
data_missing = {"x": [1, None, 3], "y": [None, 5, 6]}
df_missing = pd.DataFrame(data_missing)
print("\n1. 含缺失值的原始DataFrame:")
print(df_missing)

print("\n- 检测缺失值（True表示缺失）:")
print(df_missing.isnull())

print("\n- 每列缺失值数量统计:")
print(df_missing.isnull().sum())
```

2. 用 0 填充缺失值；删除含缺失值的行。

```python
# 2. 用 0 填充缺失值；删除含缺失值的行
print("\n2. 缺失值处理结果:")
print("\n- 用0填充缺失值:")
df_filled = df_missing.fillna(0)
print(df_filled)

print("\n- 删除含缺失值的行:")
df_dropped = df_missing.dropna()
print(df_dropped)
```

### 四、分组聚合

1. 按 gender 分组，求 score 的均值和最大值。

2. 按 gender 分组，统计每组人数。

```python
import pandas as pd

# 直接在当前单元格定义所需数据（无需依赖前面的代码）
df_group = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})

print("="*50)
print("四、分组聚合")
print("="*50)

# 1. 按 gender 分组，求 score 的均值和最大值
print("\n1. 按gender分组，score的均值和最大值:")
grouped_stats = df_group.groupby("gender")["score"].agg(["mean", "max"])
print(grouped_stats)

# 2. 按 gender 分组，统计每组人数
print("\n2. 按gender分组，每组人数统计:")
grouped_count = df_group.groupby("gender").size()
print(grouped_count)
```

### 五、排序与去重

1. 按 score 升序排序。

2. 给 DataFrame 添加一行重复数据（如 Tom,85, 男），去重。

```python
import pandas as pd

# 直接在当前单元格定义所需数据
original_df = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})

print("="*50)
print("五、排序与去重")
print("="*50)

# 1. 按 score 升序排序
print("\n1. 按score升序排序:")
df_sorted = original_df.sort_values("score", ascending=True)
print(df_sorted)

# 2. 添加一行重复数据并去重（Pandas 2.0+ 使用concat替代append）
print("\n2. 重复数据处理:")
print("\n- 添加重复数据后的DataFrame:")
# 创建重复行的DataFrame
duplicate_row = pd.DataFrame([["Tom", 85, "男"]], columns=original_df.columns)
# 纵向拼接并重置索引
df_with_dup = pd.concat([original_df, duplicate_row], ignore_index=True)
print(df_with_dup)

print("\n- 去重后的DataFrame:")
df_dedup = df_with_dup.drop_duplicates()
print(df_dedup)
```

### 六、合并与拼接

1. 横向合并：

   df1 = pd.DataFrame ({"id":[1,2], "name":["X","Y"]})

   df2 = pd.DataFrame ({"id":[1,2], "age":[20,30]})

   按 id 合并。

```python
print("\n" + "="*50)
print("六、合并与拼接")
print("="*50)

# 1. 横向合并（按共同列id合并）
merge_df1 = pd.DataFrame({"id": [1, 2], "name": ["X", "Y"]})
merge_df2 = pd.DataFrame({"id": [1, 2], "age": [20, 30]})

print("\n1. 按id横向合并:")
print("\n- 第一个DataFrame:")
print(merge_df1)
print("\n- 第二个DataFrame:")
print(merge_df2)

# 内连接（默认），保留两边都存在的id
df_merged = pd.merge(merge_df1, merge_df2, on="id")
print("\n- 合并结果:")
print(df_merged)
```

2. 纵向拼接：

   df_a = pd.DataFrame ({"val":[1,2]})

   df_b = pd.DataFrame ({"val":[3,4]})

   拼接后重置索引。

```python
import pandas as pd  # 这行是关键，必须添加

# 2. 纵向拼接（按行堆叠）
concat_df1 = pd.DataFrame({"val": [1, 2]})
concat_df2 = pd.DataFrame({"val": [3, 4]})

print("\n2. 纵向拼接并重置索引:")
print("\n- 第一个DataFrame:")
print(concat_df1)
print("\n- 第二个DataFrame:")
print(concat_df2)

# 拼接后丢弃原有索引，生成新的连续索引
df_concat = pd.concat([concat_df1, concat_df2], ignore_index=True)
print("\n- 拼接结果:")
print(df_concat)
```

### 七、时间处理

1. 创建时间列：["2025-01-01","2025-01-02","2025-01-03"]，转为 datetime 格式。

```python
# ====================== 七、时间处理 ======================
print("\n" + "="*50)
print("七、时间处理")
print("="*50)

# 1. 创建时间列并转为datetime格式
dates = ["2025-01-01", "2025-01-02", "2025-01-03"]
df_time = pd.DataFrame({"date_str": dates})

print("\n1. 字符串转datetime格式:")
df_time["date"] = pd.to_datetime(df_time["date_str"])
print(df_time)
print("\n- 数据类型:")
print(df_time.dtypes)
```

2. 提取时间列的月份、星期。

```python
# 2. 提取时间列的月份和星期
print("\n2. 提取月份和星期:")
# 提取月份（数字1-12）
df_time["month"] = df_time["date"].dt.month
# 提取星期名称（英文）
df_time["weekday_name"] = df_time["date"].dt.day_name()
# 提取星期数字（0=周一，6=周日）
df_time["weekday_num"] = df_time["date"].dt.weekday

print(df_time)
```

### 八、高级操作

1. 对 score 列做标准化（(值 - 均值)/ 标准差）。

```python
# ====================== 八、高级操作 ======================
print("\n" + "="*50)
print("八、高级操作")
print("="*50)

df_advanced = original_df.copy()

# 1. 对 score 列做标准化（Z-score标准化：(值 - 均值)/标准差）
print("\n1. score列标准化结果:")
# 计算均值和标准差
score_mean = df_advanced["score"].mean()
score_std = df_advanced["score"].std()
# 应用标准化公式
df_advanced["score_std"] = (df_advanced["score"] - score_mean) / score_std
print(df_advanced[["name", "score", "score_std"]])
print(f"\n- 原始均值: {score_mean:.2f}，原始标准差: {score_std:.2f}")
print(f"- 标准化后均值: {df_advanced['score_std'].mean():.2f}，标准化后标准差: {df_advanced['score_std'].std():.2f}")
```

2. 用透视表：行是 gender，值是 score，聚合函数为中位数。

```python
# 2. 透视表：行是gender，值是score，聚合函数为中位数
print("\n2. 按gender分组的score中位数透视表:")
pivot_table = pd.pivot_table(
    df_advanced,
    index="gender",  # 行分组字段
    values="score",  # 要聚合的数值列
    aggfunc="median"  # 聚合函数：中位数
)
print(pivot_table)
```
